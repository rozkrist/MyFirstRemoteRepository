STORE THE DATA TO MYSQL DATABASE

```python
# Check connection
import mysql.connector

# Establishing a connection to the database
cnx = mysql.connector.connect(user="root",
                              password= "",
                              host="localhost",
                              database="unirankings")
```

```python
# Uploading created Excel files to respective tabels to store the data
import pandas as pd
import pymysql
from sqlalchemy import create_engine

engine = create_engine("mysql+pymysql://root:@localhost/unirankings")

universities = pd.read_excel(r"C:\Users\krent\OneDrive\Desktop\excel for db\universities.xlsx")
universities.to_sql(name="universities",con=engine,if_exists="append",index=False)
```
